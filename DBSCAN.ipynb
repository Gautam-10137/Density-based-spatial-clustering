# Here we are comparing K-means clustering and DBSCAN on variety of data sets

```python
moons=pd.read_csv("cluster_moons.csv")
```

```python
moons.head()
```

```python
sns.scatterplot(data=moons,x='X1',y='X2')
```

```python
circles=pd.read_csv("cluster_circles.csv")
```

```python
circles.head()
```

```python
sns.scatterplot(data=circles,x='X1',y='X2')
```

```python
# Now visualizing the outputs
def display_categories(model,data):
    labels=model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3,n_init='auto')
```

```python
display_categories(model,blobs)
```

```python
# moons data
model=KMeans(n_clusters=2,n_init='auto')
```

# KMeans is using distance metrics so it doesn't perform well in moons data

```python
display_categories(model,moons)
```

```python
#  circle data
display_categories(model,circles)
```

```python
#  here also KMeans does't perform well. It will be complex to find clusters with large complex datasets.
```

```python
from sklearn.cluster import DBSCAN
```

```python
model=DBSCAN()
```

```python
display_categories(model,blobs)     # also displays outliers
```

```python
#  moons data
model=DBSCAN(eps=0.15)
display_categories(model,moons)
```

```python
#  circles data
display_categories(model,circles)
```
